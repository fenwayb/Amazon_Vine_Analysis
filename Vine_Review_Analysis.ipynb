```python
import pandas as pd
```

```python
columns = ["review_id", "star_rating", "helpful_votes", "total_votes", "vine", "verified purchase"]
vine_df = pd.read_csv("vine_table.csv", header=None)
vine_df.columns = columns
vine_df
```

```python
vine_df.dtypes
```

```python
total_votes_df = vine_df[vine_df["total_votes"] >= 20]
total_votes_df
```

```python
helpful_df = total_votes_df[total_votes_df["helpful_votes"] / total_votes_df["total_votes"] >= .5]
helpful_df
```

```python
paid_df = helpful_df[helpful_df["vine"] == "Y"]
paid_df
```

```python
free_df = helpful_df[helpful_df["vine"] == "N"]
free_df
```

```python
paid_total = len(paid_df["review_id"])
paid_five = len(paid_df[paid_df["star_rating"] == 5])
paid_per = paid_five/paid_total
paid_per
```

```python
free_total = len(free_df["review_id"])
free_five = len(free_df[free_df["star_rating"] == 5])
free_per = free_five/free_total
free_per
```
